```python
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
# to draw pictures in jupyter notebook
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

```python
data.info()
```

**1. How many men and women (*sex* feature) are represented in this dataset?**

**1。 此数据集中有多少男女（*性别*特征）？**

```python
data.groupby('sex')['sex'].count()
```

**2. What is the average age (*age* feature) of women?**

**2。 女性的平均年龄（*年龄*特征）是多少？**

```python
data[data['sex']=='Female']['age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

3.德国公民的百分比（原籍国特色）是多少？

```python
data['native-country'].value_counts(normalize=True)['Germany']*100
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?**

**4-5。 年收入超过50K（*工资*特征）和年收入低于50K的人的年龄平均值和标准差是多少？**

```python
data.groupby(['salary'])[['age']].describe()
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

**6。 收入超过50K的人是否至少接受过高中教育？ （*教育 - 学士，教授，副学士，副教授，硕士*或*博士*专业）**

```python
data[data['salary']=='>50K'].groupby('education').count()
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

**7。 显示每个种族（*种族*特征）和每个性别（*性别*特征）的年龄统计数据。 使用* groupby（）*和* describe（）*。 找出* Amer-Indian-Eskimo *种族的最大年龄。**

```python
data.groupby(['race','sex'])['age'].describe()
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

**8。 其中挣多少（> 50K）的人比例：已婚或单身男性（*婚姻状况*特征）？ 考虑结婚那些有*婚姻状况的人*以*已婚*（已婚 - 公民 - 配偶，已婚 - 配偶 - 缺席或已婚 - AF-配偶）开始，其余的被视为单身汉。**

```python
data['Married']=data['marital-status'].map({'Married-civ-spouse':True,'Married-spouse-absent':True,'Married-AF-spouse':True})
data['Married']=data['Married'].fillna(False)
data.head()
```

```python
data[data['salary']=='>50K']['Married'].value_counts(normalize=True)
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

**9。 一个人每周工作的最长小时数（每周*小时*功能）是多少？ 有多少人工作了这么多小时，那些赚了很多（> 50K）的人的百分比是多少？**

```python
data['hours-per-week'].max()
```

```python
data[data['hours-per-week']==data['hours-per-week'].max()].shape[0]
```

```python
data[data['hours-per-week']==data['hours-per-week'].max()]['salary'].value_counts(normalize=True)
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

**10。 计算每个国家（*本国*）赚取少量和大量（*工资*）的人的平均工作时间（*每周*小时*）。 这对日本会是什么？**

```python
data.groupby(['native-country','salary'])['hours-per-week'].agg([np.mean])
```

```python
data[data['native-country']=='Japan'].groupby(['salary'])['hours-per-week'].agg([np.mean])
```
